# Tests with autograd

```python
# import gridworld library - make sure this is executed prior to running any gridworld cell
import sys
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

sys.path.append('demo_python_backend_files')
%matplotlib notebook
%load_ext autoreload
%autoreload 2
```

```python
# load up enviro
from gridworld_lib import openai_gym_enviro
environment = openai_gym_enviro.environment(environment = 'cartpole',training_episodes = 4000,validation_episodes = 50)
```

```python
# initialize learner
from gridworld_lib import openai_gym_qlearner
qlearner = openai_gym_qlearner.learner(environment = environment)
```

```python
# run q-learning
qlearner.train(validate = False,alpha = 10**-3,action_method = 'random',gamma = 0.99,training_episodes = 200,validation_episodes = 50,max_steps = 100)
```

```python
### plot average rewards ###
# create figure
fig = plt.figure(figsize = (12,5))
ax = fig.add_subplot(1,1,1)

# plot total reward history
ax.plot(qlearner.training_reward)
ax.plot(qlearner.validation_reward)
ax.legend(['training','validation'],loc='center right', bbox_to_anchor=(1, .5))
ax.set_xlabel('episode')
ax.set_ylabel('total reward')
plt.show()
```
